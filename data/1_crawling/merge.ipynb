```python
import pandas as pd
import os

import os

path = './card_data/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')]


benefit_path = './benefit_data/'
benefit_file_list = os.listdir(benefit_path)
benefit_file_list_py = [file for file in benefit_file_list if file.endswith('.csv')]
```

```python
card = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i, header=None, encoding='cp949')
    card = pd.concat([card,data])
    
card = card.reset_index(drop = True)
```

```python
benefit = pd.DataFrame()
for i in benefit_file_list_py:
    benefit_data = pd.read_csv(benefit_path + i, header=None, encoding='cp949')
    benefit = pd.concat([benefit,benefit_data])
    
benefit = benefit.reset_index(drop = True)
```
