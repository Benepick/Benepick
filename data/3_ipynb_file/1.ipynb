```python
import pandas as pd
```

```python
card = pd.read_csv('card_data.csv', header=None, encoding="cp949")
benefit = pd.read_csv('card_benefit.csv', encoding="cp949")
```

```python
card.columns = ["card_id", "card_name", "card_brand", "card_fee_in", 'card_record', "card_type", "card_image"]
```

```python
card.head()
```

```python
card['domestic_fee'] = card['card_fee_in'].str.extract(r'국내전용 (\d+,?\d+)원')
card['overseas_fee'] = card['card_fee_in'].str.extract(r'해외겸용 (\d+,?\d+)원')
```

```python
card['domestic_fee'].fillna(0, inplace=True)
card['overseas_fee'].fillna(0, inplace=True)
```

```python
card['card_record'] = card['card_record'].str.replace('전월실적', '')
```

```python
del card["card_fee_in"]
```

```python
card
```

```python
def extract_card_types(card_type_string):
    if isinstance(card_type_string, str):  # nan 처리를 위한 조건 추가
        card_types = []
        if 'mastercard' in card_type_string.lower():
            card_types.append('mastercard')
        if 'visa' in card_type_string.lower():
            card_types.append('VISA')
        if 'amex' in card_type_string.lower():
            card_types.append('AMEX')
        if 'unionpay' in card_type_string.lower():
            card_types.append('UnionPay')
        if 'jcb' in card_type_string.lower():
            card_types.append('JCB')
        return card_types
    else:
        return []

card['extracted_card_types'] = card['card_type'].apply(extract_card_types)
```

```python
del card["card_type"]
```
