```python
import pandas as pd
```

```python
benefit = pd.read_csv('card_benefit.csv', header=None, encoding="cp949")
```

```python
benefit.columns = ["card_id", "card_name", "card_brand", "content", 'detail']
```

```python
benefit
```

```python
def split_detail(detail):
    return detail.split('\r\n')

# 새로운 열에 분리된 내용 저장
benefit['split_detail'] = benefit['detail'].apply(split_detail)
```

```python
del benefit['detail']
```

```python
benefit
```

```python
for index, detail in enumerate(benefit['split_detail']):
    for d_detail in detail:
        if "전월실적" in d_detail:
            benefit.at[index, "pre_result"] = d_detail
```
