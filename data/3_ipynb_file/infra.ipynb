## 주유소 처리

```python
import pandas as pd

infra = pd.read_csv('processed_local.csv')
```

```python
infra.loc[(infra['개방서비스명'].str.contains('주유소')) | (infra['개방서비스명'].str.contains('석유')), '개방서비스명']='주유소'
infra.loc[(infra['개방서비스명'].str.contains('주유소')) | (infra['개방서비스명'].str.contains('석유')), '업태구분명']='주유소'
```

```python
infra.loc[(infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('GS칼텍스')) | ((infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('지에스칼텍스'))), '소분류'] ='GS'
infra.loc[(infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('SK에너지')) | ((infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('에스케이'))), '소분류'] ='SK'
infra.loc[(infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('현대오일')), '소분류'] ='현대'
infra.loc[(infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('S-OIL')), '소분류'] ='S-OIL'
infra.loc[(infra['개방서비스명']=='주유소') & (infra['사업장명'].str.contains('알뜰')), '소분류'] ='알뜰'
```

```python
infra[(infra['개방서비스명']=='주유소')].to_csv("주유소.csv", index=False, encoding="utf-8-sig")
```

```python
infra.loc[(infra['개방서비스명'] == '대규모점포') & (infra['사업장명'].str.contains('신세계')) & (infra['업태구분명'].str.contains('전문점')), '업태구분명']='아울렛'
```
